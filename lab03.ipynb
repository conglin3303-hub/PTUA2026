```python
cover_price = 24.95
discount = 0.4
copies = 60
shipping_first = 3
shipping_additional = 0.75

unit_cost = cover_price * (1 - discount)

total_book_cost = copies * unit_cost

total_shipping = shipping_first + shipping_additional * (copies - 1)

total_wholesale_cost = total_book_cost + total_shipping


print(f"The total wholesale cost for 60 copies is：${total_wholesale_cost:.2f}")
```

```python
def first(word):
    return word[0]


def last(word):
    return word[-1]


def middle(word):
    return word[1:-1]

def is_palindrome(word):
    if len(word) <= 1:
        return True
    if first(word) != last(word):
        return False
    return is_palindrome(middle(word))


print(is_palindrome('allen'))
print(is_palindrome('bob'))
print(is_palindrome('otto'))
print(is_palindrome('redivider'))
```

```python
def rotate_letter(letter, n):
    if letter.isupper():
        start = ord('A')
    elif letter.islower():
        start = ord('a')
    else:
        return letter

    c = ord(letter) - start
    i = (c + n) % 26 + start
    return chr(i)

def rotate_word(word, n):
      res = ""
      for letter in word:
        res += rotate_letter(letter, n)
        return res

if __name__ == '__main__':
    print(rotate_word('cheer', 7))
    print(rotate_word('melon', 10))
    print(rotate_word('sleep', 9))
```
